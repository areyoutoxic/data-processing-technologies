### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
import requests
from bs4 import BeautifulSoup

with open('C:\\Users\\marya\\Downloads\\ТОД\\3 лаба\\data\\steps_sample.xml') as f:
    ab = BeautifulSoup(f, 'xml')
````

````python
res = list();
for recipe in ab.recipes.find_all("recipe"):
    st = [steps.next for steps in recipe.steps.find_all("step")]
    res.append({recipe.find("id").next: st})
````

````python
import json 
with open('C:\\Users\\marya\\Downloads\\ТОД\\3 лаба\\data\\steps_sample.json', mode='w', encoding='utf-8') as f: # открываем файл на запись
    json.dump(res, f, indent=2)
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
unique_count_steps = []
for i in res:
    for key in i:
        if len(i[key]) not in unique_count_steps:
            unique_count_steps.append(len(i[key]))
````

````python
dict_2 = dict.fromkeys(unique_count_steps, [])
````

````python
for i in res:
    for key in i:
        #print(key)#идентификатор 
        l = len(i[key])
        dict_2[l] = dict_2.get(l, []) + [key]
dict_2
````
